```python
new_df=df.copy()
```

```python
new_df.TotalCharges=pd.to_numeric(new_df.TotalCharges,errors='coerce')
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
new_df.loc[new_df['TotalCharges'].isnull()==True]
```

```python
new_df.dropna(how='any',inplace=True)
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
labels=['1-12','13-24','25-36','37-48','49-60','61-72']
bins=[0,12,24,36,48,60,71]

new_df['tenure']=pd.cut(new_df['tenure'],labels=labels,bins=bins,include_lowest=True)
```

```python
new_df['tenure'].value_counts()
```

```python
new_df.drop(columns=['customerID'],axis=1,inplace=True)
```
